```python
import string

# 英単語を格納するリスト
words = []

# アルファベットごとの数を格納する辞書（a〜zを0で初期化）
letter_counts = {ch: 0 for ch in string.ascii_lowercase}

# 入力受付（未入力で終了）
while True:
    w = input('英単語を入力してください：')
    if w == "":
        break
    
    # 単語はそのまま保存（表示用）
    words.append(w)
    
    # 出現回数は小文字化して a〜z だけ数える
    for ch in w.lower():
        if ch in letter_counts:
            letter_counts[ch] += 1

# ABC順に並べ替え（大文字小文字を区別せずにソート）
sorted_words = sorted(words, key=str.lower)
print(f'入力した英単語：{sorted_words}')

# 0より大きい文字だけ表示
for ch in string.ascii_lowercase:
    if letter_counts[ch] > 0:
        print(f'{ch}が{letter_counts[ch]}個ありました')
```
